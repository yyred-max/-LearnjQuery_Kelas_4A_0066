1. METODE GRAFIK


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.1)
f = lambda x:x**3+x**2-3*x-3
plt.plot(x,f(x), 'brown')
```

2. METODE TABULASI


```python
x=np.arange(-5,5,1)
print("x:",x)
print("y:",f(x))
print("Akar persamaan dari f(x)=-1")
```
